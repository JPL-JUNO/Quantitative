```python
import numpy as np
import matplotlib.pyplot as plt

iv = np.arange(-4, 4.1, .1)
pv = np.arange(-3, 3.1, .1)

DELTA = 0
GAMMA = -10029
VEGA = -56951
THETA = 25541
UNDERLYING = 0

xx, yy = np.meshgrid(iv, pv)

# zz = yy * GAMMA + xx ** 2 * VEGA + THETA + DELTA + UNDERLYING
zz = yy ** 2 * GAMMA + xx * VEGA + THETA + DELTA + UNDERLYING

cs = np.sort(np.append(0, np.linspace(np.min(zz), np.max(zz), 10)))
fig, ax = plt.subplots(figsize=(8, 6))
cset = ax.contourf(xx, yy, zz, alpha=.5)
contour = ax.contour(
    xx, yy, zz, colors='k')
ax.set_xlabel('IV')
ax.set_ylabel('PV')
ax.clabel(contour, colors='r')
ax.plot([2.58], [-2.38], 'r*', markersize=10)
plt.colorbar(cset)
plt.tight_layout()
```
